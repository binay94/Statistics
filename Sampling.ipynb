```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/sunnysavita10/Statistics_With_Python/main/Iris.csv")
```

```python
data
```

```python
#Random Sampling
random_sample=data.sample(10)
```

```python
random_sample
```

```python
#Systematic Sampling
sys_sample=data.iloc[10:30:2]
```

```python
sys_sample
```

```python
#Cluster Sampling
clusters = np.random.choice(np.arange(1,151), size=6, replace=False)
```

```python
clusters
```

```python
cluster_sample = data[data['Id'].isin(clusters)]
```

```python
cluster_sample
```

```python
#stratified Sampling
startified_sample = data.groupby('Species', group_keys=False).apply(lambda x: x.sample(2))
```

```python
startified_sample
```
